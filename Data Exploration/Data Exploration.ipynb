# Utils

```python
def segment(img, roi):
    img[:, :, 0] = img[:, :, 0]*roi
    img[:, :, 1] = img[:, :, 1]*roi
    img[:, :, 2] = img[:, :, 2]*roi
    return img
```

```python
def visualize(imgs, rois, rows, columns, width, height, ex_count, title, font_size, RGB = False):
    fig = plt.figure(figsize=(width, height))
    index = 1

    for x in range(5):
        img = mpimg.imread(imgs[x])
        if RGB:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        fig.add_subplot(rows, columns, index)
        plt.imshow(img
#                    , cmap = 'gist_rainbow'
                  )
        plt.axis('off')
        plt.title(str(img.shape) + f' max: {round(img.max(), 2)}' + f' min: {round(img.min(), 2)}')
        index += 1
        roi = mpimg.imread(rois[x])
        fig.add_subplot(rows, columns, index)
        plt.imshow(roi, cmap = 'gray')
        plt.axis('off')
        plt.title(str(roi.shape) + f' max: {round(roi.max(), 2)}' + f' min: {round(roi.min(), 2)}')
        index += 1
        seg = segment(img, roi)
        fig.add_subplot(rows, columns, index)
        plt.imshow(seg
#                    , cmap = 'gist_rainbow'
                  )
        plt.axis('off')
        plt.title(str(seg.shape) + f' max: {round(seg.max(), 2)}' + f' min: {round(seg.min(), 2)}')
        index += 1
    fig.suptitle(title, fontsize=font_size)
```

# Training Analysis

```python
visualize(imgs = train_nor_img_files, 
          rois = train_nor_roi_files, 
          rows = 5, 
          columns = 3, 
          width = 10 , 
          height = 10, 
          ex_count = 5,
          title = 'Training Normal IMG vs ROI', 
          font_size = 16,
          RGB = False)
```
