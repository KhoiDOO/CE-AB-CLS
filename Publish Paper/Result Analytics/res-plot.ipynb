```python
import seaborn as sns
sns.set_theme(style="ticks")

dots = sns.load_dataset("dots")

# Define the palette as a list to specify exact values
palette = sns.color_palette("rocket_r")

# Plot the lines on two facets
sns.relplot(
    data=dots,
    x="time", y="firing_rate",
    hue="coherence", size="choice", col="align",
    kind="scatter", size_order=["T1", "T2"], palette=palette,
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="white", context="talk")
rs = np.random.RandomState(8)

# Set up the matplotlib figure
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(7, 5), sharex=True)

# Generate some sequential data
x = np.array(list("ABCDEFGHIJ"))
print(x)
y1 = np.arange(1, 11)
sns.barplot(x=x, y=y1, palette="rocket", ax=ax1)
ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Sequential")

# Center the data to make it diverging
y2 = y1 - 5.5
sns.barplot(x=x, y=y2, palette="vlag", ax=ax2)
ax2.axhline(0, color="k", clip_on=False)
ax2.set_ylabel("Diverging")

# Randomly reorder the data to make it qualitative
y3 = rs.choice(y1, len(y1), replace=False)
sns.barplot(x=x, y=y3, palette="deep", ax=ax3)
ax3.axhline(0, color="k", clip_on=False)
ax3.set_ylabel("Qualitative")

# Finalize the plot
sns.despine(bottom=True)
plt.setp(f.axes, yticks=[])
plt.tight_layout(h_pad=2)
```

```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(os.getcwd())
```

```python
file_path = os.getcwd() + "/10fold91/re1.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df = df .drop([
    "Unnamed: 0", "Max Accuracy", "Max Precision", 
    "Max Sensitivity", "Max Specificity", "Max F1-score", "Max AUC"], axis = 1)
df.head()
```

```python
mean_df = df.iloc[:, :9]
mean_df.head()
```

```python
mean_df["Algo-K"] = [f"{x}-{y}" for x, y in zip(mean_df["Feature Descriptor"], mean_df["k-cluster"])]
mean_df.head()
```

```python
mean_df.shape
```

```python
mean_df["Method"].value_counts()
```

```python
svm_df = mean_df[mean_df["Method"] == "ExtraTreeClassifier"].drop(["Feature Descriptor", "k-cluster"], axis = 1)
svm_df.head()
```

```python
sns.set(font_scale=0.7)

f, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(6, 1, figsize=(10, 7), sharex=True)

sns.barplot(x=svm_df["Algo-K"].values, y=svm_df["Mean Accuracy"].values, palette="mako", ax=ax1)
ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Accuracy")

sns.barplot(x=svm_df["Algo-K"].values, y=svm_df["Mean Precision"].values, palette="mako", ax=ax2)
ax2.axhline(0, color="k", clip_on=False)
ax2.set_ylabel("Precision")

sns.barplot(x=svm_df["Algo-K"].values, y=svm_df["Mean Sensitivity"].values, palette="mako", ax=ax3)
ax3.axhline(0, color="k", clip_on=False)
ax3.set_ylabel("Sensitivity")

sns.barplot(x=svm_df["Algo-K"].values, y=svm_df["Mean Specificity"].values, palette="mako", ax=ax4)
ax4.axhline(0, color="k", clip_on=False)
ax4.set_ylabel("Specificity")

sns.barplot(x=svm_df["Algo-K"].values, y=svm_df["Mean F1-score"].values, palette="mako", ax=ax5)
ax5.axhline(0, color="k", clip_on=False)
ax5.set_ylabel("F1-score")

sns.barplot(x=svm_df["Algo-K"].values, y=svm_df["Mean AUC"].values, palette="mako", ax=ax6)
ax6.axhline(0, color="k", clip_on=False)
ax6.set_ylabel("AUC")

plt.xticks(rotation=90)
```
