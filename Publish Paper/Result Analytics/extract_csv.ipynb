```python
lst = ["ACC", "PRE", "SENS", "SPEC", "F1", "AUC"]
lst_drop = []
cols = df_nonmax.columns
count = 0
for col in range(1, len(cols)-1, 2):
    print(col)
    col1, col2 = cols[col], cols[col+1]
    print(col1, col2)
    df_nonmax.insert(1, lst[count], [f"({x} - {y})" for x, y in zip(df_nonmax[col1], df_nonmax[col2])], True)
    lst_drop.append(col1)
    lst_drop.append(col2)
    count += 1
df_nonmax.head()
```
